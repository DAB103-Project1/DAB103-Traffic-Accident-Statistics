# Traffic Statistics

### Data Source

(https://roadtraffic.dft.gov.uk/custom-downloads/road-accidents/)


```python
#import the useful libraries

import numpy as np
import pandas as pd
import seaborn as sns
sns.set_style("darkgrid")
import plotly.express as px
import matplotlib.pyplot as plt
import matplotlib.pyplot as plot
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
import os
import time
```

# Display the Dataset Files - Countrywise Data :-

```python
#Loading & fetching Main Dataset File - Traffic Statistics -  Countrywise
df = pd.read_excel("Traffic Statistics.xlsx")
df
```

```python
#Loading & fetching Second Dataset File - Monthwise Accidents & RoadClass
df2 = pd.read_excel("RoadClass_AccidentMonth.xlsx")
df2
```

```python
# Loading & fetching Third Dataset File - Speedlimit by JunctionType
df3 = pd.read_excel("Speedlimit_JunctionType.xlsx")
df3
```

```python
# Loading & fetching Fourth Dataset File - Accidents by Roadtype, Month & Day
df4 = pd.read_excel("Roadtype_Month_day.xlsx")
df4
```

```python
#Loading & fetching Fifth Dataset File - Accidents by LightCondition & RoadSurface
df5 = pd.read_excel("LightCondition_RoadSurface.xlsx")
df5
```

# Data Transformation :-

## Merging of Multiple Data Files into main sheet - Countrywise Data :-

```python
# Merging Second datafile with the Main Dataset File - Traffic Statistics
df_combine = df.merge(df2, how = 'left')
df_combine
```

```python
# Merging Third datafile with the Main Dataset File - Traffic Statistics
df_combine = df_combine.merge(df3, how = 'left')
df_combine
```

```python
# Merging Fourth datafile with the Main Dataset File - Traffic Statistics
df_combine = df_combine.merge(df4, how = 'left')
df_combine
```

### Main Datasheet 1 after merging all files - Traffic Statistics_Accident_Countrywise :-

```python
# Merging Fifth datafile with the Main Dataset File - Traffic Statistics 
df_combine = df_combine.merge(df5, how = 'inner')
df_combine
```

## Display the Dataset Files - Regionwise Data :-

```python
#Loading & fetching Main Dataset File - Traffic Statistics -  Regionwise
df7 = pd.read_excel("Traffic Statistics.xlsx", sheet_name = "Regions_by_severity")
df7
```

```python
# Loading & fetching Second Dataset File - Accidents by Region
df8 = pd.read_excel("Regionwise_Accidents.xlsx")
df8
```

```python
#Loading & fetching Main Dataset File - Regionwise_Casualities
df9 = pd.read_excel("Regionwise_Casualities.xlsx")
df9
```

```python
#Loading & fetching second Dataset File - Regionwise Monthly Casualities
df10 = pd.read_excel("Regionwise_Casualities_Monthly_Daily.xlsx")
df10
```

## Merging of Multiple Data Files into main sheet - Regionwise Data :-

### Main Datasheet 2 after merging all files - Traffic Statistics_Accidents_Regionwise :-

```python
#Merging Second datafile with the Main Dataset File - Traffic Statistics - (Sheet:Regions_by_serverity)
df11_combine = df7.merge(df8, how = 'left')
df11_combine
```

### Main Datasheet 3 after merging all files - Casualities_Regionwise :-

```python
#Merging Third datafile with the Main Dataset File - Casualities_Regionwise
df12_combine = df9.merge(df10, how = 'left')
df12_combine
```

### Renaming Column :-

```python
#Renaming column "Ons code" to "Unquie ID" for Main Datasheet 1
df_combine.rename(columns = {'Ons code':'Unique ID'}, inplace = True)
df_combine
```

```python
#Renaming column "Ons code" to "Unquie ID" for Main Datasheet 2
df11_combine.rename(columns = {'Ons code':'Unique ID'}, inplace = True)
df11_combine
```

```python
#Renaming column "Ons code" to "Unquie ID" for Main Datasheet 3
df12_combine.rename(columns = {'Ons code':'Unique ID'}, inplace = True)
df12_combine
```

```python
#Removing Null values from the Main Datasheet 3
#Below is the cleaned Main Datasheet 3 after droping "Nulls"
df14=df12_combine.dropna(subset=['Accident month','Accident day', 'Accident hour'])
df14
```

# Exploratory Data Analysis (EDA):-

### Analysis & Visualizations :-

```python
#Bar plot showing Monthwise Accidents - Data is being fetched from cleaned Main Datasheet 2(df13)
plt.figure(figsize=(12,7))
plt.bar(df13['Accident month'],df13['Accidents'],color='#585973')

plt.title('Monthly breakdown of Accidents')
plt.xlabel('Months')
plt.ylabel('Accidents')
plt.show()
```

```python
# Groupby function used on Speedlimit to show the total Accidents count - Data is being fetched from cleaned Main Datasheet 1(df6)
g3 = df6.groupby(['Speed limit'])['Accidents'].count()
g3
```

```python
#Bar plot shows total number of Accidents based on Severity
g3[:8].plot(kind = 'bar' , ylabel='Total Accidents', title='Speedlimit where Accidents occured',color ='#349484')
```

```python
#Pie chart showing percentage ratio of Casuality severity - Data is being fetched from cleaned Main Datasheet 3(df14)
killed = df14.loc[df14['Casualty severity'] == 'Killed'].count()[0]
killed
serious = df14.loc[df14['Casualty severity'] == 'Seriously injured'].count()[0]

slight = df14.loc[df14['Casualty severity'] == 'Slightly injured'].count()[0]

labels = ['killed', 'serious','slight']
colors = ['#abcdef', '#aabbcc','#543c59']
explode = (0,0,.1)

plt.figure(figsize=(12,10))
plt.pie([killed,serious,slight],explode = explode,labels = labels,colors=colors, autopct='%.2f %%')
plt.title('Casualty severity by percentage')
plt.show()
```

```python
# Groupby function used on Accident day & Accident Severity to show the total Accidents count - Data is being fetched from cleaned Main Datasheet 1(df6)
g4 = df6.groupby(['Accident day','Accident severity'])['Accidents'].count()
g4
```

```python
# Bar plot shows breadown of Accidents & their severity each day
g4[:20].plot(figsize=(12,12), kind = 'barh' , ylabel='Daily accidents of varying severity', title='Statistic on accident severity and frequency by day',color ='#543c59')
```

```python
# Groupby function shows various Road types in each - Region Data is being fetched from cleaned Main Datasheet 2(df13)
g5 = df13.groupby(['Region','Road type'])['Region'].count()
g5
```

```python
g5[:23].plot(figsize=(12,12), kind = 'barh' , ylabel='Regional Road types', title='Road type breakdown by region',color ='#abcdef')
```

```python
##Bar plot shows total number of Accidents per Region - Data is being fetched from cleaned Main Datasheet 2(df13)
plt.figure(figsize=(10,5))
plt.title("Accidents per Region")
plt.ylabel("Region")
sns.set_palette("autumn_r")
ax= sns.countplot(data=df13, y="Region")
for p in ax.patches: 
    width = p.get_width()
    plt.text(2+p.get_width(), p.get_y()+0.5*p.get_height(),
             '{:1.0f}'.format(width),
             ha='center', va='center')
```

```python
# Groupby function shows Casualties by count age - Data is being fetched from cleaned Main Datasheet 3(df14)
g6 = df14.groupby(['Casualty sex'])['Casualty age'].count()
g6
```

```python
g6[:6].plot(figsize=(12,12), kind = 'barh' , ylabel='Count', title='Casualties by count age',color ='#abcdef')
```

```python
#Accident count across each Region - Data is being fetched from cleaned Main Datasheet 2(df13)
g7 = df13.groupby(['Region']).size();
g7
```

```python
##Line graph shows Accident count based on Region 
p2 = g7.plot(kind='line', title='Accident Count',x='Region')
p2
```

```python
#Bar graph representing various categories of Weather Conditions causing Accidents
g8 = df14.groupby(['Detailed road user', 'Region'])['Region'].count()
g8
```

```python
g8[:6].plot(figsize=(12,12), kind = 'barh' , ylabel='Count', title='Road User breakdown by Region',color ='#aabbcc')
```
